```python
from qiskit import qasm2

program = '''
    OPENQASM 2.0;
    include "qelib1.inc";
    qreg q[5];
    creg c[5];

    cx q[0], q[1];
    cx q[2], q[4];
    cx q[1], q[3];
    h q[3];
    cx q[1], q[2];
    cx q[0], q[2];
    cx q[0], q[1];
    cx q[2], q[3];
    cx q[2], q[4];
    cx q[3], q[2];
'''

# circ = qasm2.loads(program)
filepath = "./benchmark/sabre_CXonly_bench/small/alu-v0_27_example_CXonly.qasm"
filepath2 = "./benchmark/sabre_bench/small/alu-v0_27_example.qasm"
filepath3 = "./benchmark/mqtbench/qubit_50_53/ae_indep_qiskit_50.qasm"
filepath4 = "./benchmark/mqtbench/qubit_16_20/qft_indep_qiskit_20.qasm"
filepath5 = "./benchmark/mqtbench/qubit_16_20/random_indep_qiskit_20.qasm"
filepath6 = "./benchmark/mqtbench/qubit_16_20/wstate_indep_qiskit_20.qasm"

# circ = qasm2.load(filepath)
circ = qasm2.loads(program)

nQubit = len(circ.qubits)
nGate = len(circ.data)
n2QGate = 0

interactivity = [ [] for _ in range(nQubit) ]

print(f"# of logical qubits: {nQubit}")
print(f"# of total gates: {nGate}")
print(f"# of 2-qubit gates: {n2QGate}")
print("input circuit composition:", dict(circ.count_ops()))
print("circuit instructions: ")
for index, inst in enumerate(circ.data):
    if inst.operation.num_qubits == 1:
        circ.data.pop(index)
        print("g_{}: {}({})".format(
            index+1,
            inst.operation.name,
            inst.qubits[0]._index,
        ))
    elif inst.operation.num_qubits == 2:
        n2QGate += 1
        first = inst.qubits[0]._index
        second = inst.qubits[1]._index
        print("g_{}: {}({}, {})".format(
            index+1,
            inst.operation.name,
            first,
            second
        ))
        interactivity[first].append(index+1)
        interactivity[second].append(index+1)
    else:
        print("More than two operands")

print(f"CIRCUIT DEPTH: {circ.depth()}")
print("input circuit: ")
# circ.draw(output='mpl')
circ.draw(output='mpl', filename="./result/input_circ.png", scale=0.7)
```

```python
print(dag.input_map)
# input_map  key  qubitvalue  qubit  DAG  node ()
for bit, gate in dag.input_map.items():
    if isinstance(bit, Clbit):
        print(f"meet clbit: {bit}")
        print("------------------------------------------")
        continue

    if isinstance(gate, DAGInNode):
        print(f"{bit}: DAGInNode({gate.wire})")
    elif isinstance(gate, DAGOpNode):
        print(f"{bit}: DAGOpNode({gate.wire})")
    elif isinstance(gate, DAGOutNode):
        print(f"{bit}: DAGOutNode({gate.wire})")

    # edges(gate)  gate  DAG  output edges
    for _, successor, edge_data in dag.edges(gate):
        assert type(gate) is type(_), "type something wrong!"
        print(f"type of successor: {type(successor)}")
        assert type(edge_data) is Qubit, "type of DAG edge is not <Qubit>"
        print(f"    to {successor.op.name}({successor.qargs}): {edge_data}")
    
    print("------------------------------------------")
```

```python
print(len(list(dag.edges())))
def get_next_edge(dag):
    precedences = defaultdict(list[DAGOpNode])
    in_degree = defaultdict(int)
    count2 = 0
    count = 0
    for pred, succ, edge_data in dag.edges():
        if isinstance(pred, DAGInNode) and isinstance(pred.wire, Clbit) and isinstance(succ, DAGOutNode):
            print(f"DAGInNode({pred.wire}) to DAGOutNode({succ.wire}): {edge_data}")
            count2 += 1
            continue
        elif isinstance(pred, DAGInNode) and isinstance(succ, DAGOpNode):
            # if isinstance(pred.wire, Clbit):
                # if isinstance(succ, DAGOutNode):
                #     print(f"DAGInNOde({pred.wire}) to DAGOutNode({succ.wire}): {edge_data}")
                # continue
            print(f"DAGInNode({pred.wire}) to {succ.op.name}{succ.qargs}: {edge_data}")
            count2 += 1
            continue
        elif isinstance(pred, DAGOpNode) and isinstance(succ, DAGOutNode):
            print(f"{pred.op.name}{pred.qargs} to DAGOutNode({succ.wire}): {edge_data}")
            count2 += 1
            continue
        print(f"{pred.op.name}{pred.qargs} to {succ.op.name}{succ.qargs}: {edge_data}")
        count += 1
        precedences[pred].append(succ)
        in_degree[succ] += 1
        # dag.remove_op_node(pred)
        return pred, succ, edge_data
        # if count == 1:
        #     break
    print(f"count = {count}")
    print(f"count2 = {count2}")
```

```python
from qiskit.transpiler import PassManager
from qiskit.transpiler.passes import (
    Decompose,
    RemoveDiagonalGatesBeforeMeasure
)

def get_non_single_qg_names(non_single_qg):
    non_single_qg_names = list()
    for gate in non_single_qg:
        if gate.name == "cx":
            continue
        non_single_qg_names.append(gate.name)
    return list(non_single_qg_names)

two_qg_list = dag.two_qubit_ops()
mul_qg_list = dag.multi_qubit_ops()
non_single_qg_names = get_non_single_qg_names(two_qg_list + mul_qg_list)

init_pm = PassManager()
init_pm.append([
    RemoveDiagonalGatesBeforeMeasure(),
    Decompose(non_single_qg_names)
])
init_circ = init_pm.run(circ)
# init_circ.draw(output='mpl')
print("initial circuit composition:", dict(init_circ.count_ops()))
init_circ.draw(output='mpl', filename="./result/init_circ.png", scale=0.5)
```
